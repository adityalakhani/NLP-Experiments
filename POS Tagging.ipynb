```python
import nltk

text = nltk.word_tokenize("As the moonlight faded, replaced by the first rays of sunshine, the child slowly lowered his palms. His eyes opened to reveal deep, dark pupils that seemed to have a sharp golden glint deep within them. Huang Xiaolong breathed out a mouthful of foul air, his eyes staring at the rising sun. From the day he was born to the present day, it had been seven years since he came to this world. He started practicing the Body Metamorphose Scripture at the age of three. Now four years later, he had managed to reach the Third Stage: Palm Propping the Sky Gate. In his previous life, his ancestors had trained under the banner of the Shaolin Temple and the Body Metamorphose Scripture was an inheritance from his ancestors. Until now, Huang Xiaolong had been unable to understand the reason why someone like him, who was hailed as a martial arts prodigy in his time, was brought to this world. And the matter that depressed Huang Xiaolong the most was that he awoke in the body of a newborn baby")
nltk.pos_tag(text)
```
